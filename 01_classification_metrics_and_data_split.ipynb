```python
import pandas as pd



X_all = pd.DataFrame(mnist.data)
y_all = pd.Series(mnist.target.astype(int))  # zamieniamy etykiety na liczby całkowite

y_sorted = y_all.sort_values()
sorted_indices = y_sorted.index

# sortujemy dane i etykiety
X_sorted = X_all.reindex(sorted_indices).reset_index(drop=True)
y_sorted = y_sorted.reset_index(drop=True)



X_train, X_test = X_sorted[:56000], X_sorted[56000:]
y_train, y_test = y_sorted[:56000], y_sorted[56000:]

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_sorted, y_sorted, test_size=0.2, random_state=42)
print(y_train)

```

```python
print(y_test)
```

```python
import pickle
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_score

# 0 = true, inne = false
y_train_0 = (y_train == 0)
y_test_0 = (y_test == 0)

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_0)

# oblicz dokładność
acc_train = sgd_clf.score(X_train, y_train_0)
acc_test = sgd_clf.score(X_test, y_test_0)

print(acc_train, acc_test)

with open('sgd_acc.pkl', 'wb') as f:
    pickle.dump([acc_train, acc_test], f)

# validatcja krzyzowa
cv_scores = cross_val_score(sgd_clf, X_train, y_train_0, cv=3, scoring='accuracy')
print(cv_scores)

with open('sgd_cva.pkl', 'wb') as f:
    pickle.dump(cv_scores, f)

```

```python
from sklearn.metrics import confusion_matrix

sgd_clf_all = SGDClassifier(random_state=42)
sgd_clf_all.fit(X_train, y_train)

```

```python
y_pred = sgd_clf_all.predict(X_test)
print(y_pred)
```

```python

cm = confusion_matrix(y_test, y_pred)

print(cm)

with open('sgd_cmx.pkl', 'wb') as f:
    pickle.dump(cm, f)
```
